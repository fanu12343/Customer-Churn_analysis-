```python
prod_churn = df.copy()

prod_churn.fillna(
    {
        "Value_Deal": "None",
        "Multiple_Lines": "No",
        "Internet_Type": "None",
        "Online_Security": "No",
        "Online_Backup": "No",
        "Device_Protection_Plan": "No",
        "Premium_Support": "No",
        "Streaming_TV": "No",
        "Streaming_Movies": "No",
        "Streaming_Music": "No",
        "Unlimited_Data": "No",
        "Churn_Category": "Others",
        "Churn_Reason": "Others"
    },
    inplace=True
)
```

```python
prod_churn.isnull().sum()
```

```python
vw_churndata = prod_churn[
    prod_churn["Customer_Status"].isin(["Churned", "Stayed"])
]
```

```python
vw_JoinData=prod_churn[prod_churn['Customer_Status']=='Joined']
```

```python
prod_churn.to_csv('/content/drive/MyDrive/prod_churn.csv',index=False)
vw_churndata.to_csv('/content/drive/MyDrive/vw_churndata.csv',index=False)
vw_JoinData.to_csv('/content/drive/MyDrive/vw_JoinData.csv',index=False)



```
